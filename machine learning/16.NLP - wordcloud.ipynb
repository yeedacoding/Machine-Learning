# WordCloud

## 1. 이상한 나라의 앨리스

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open('datas/06_alice.txt').read()
alice_mask = np.array(Image.open('datas/06_alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
print(text)
```

```python
stopwords
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white", 
               max_words=2000,  # 표현할 최대 단어 갯수
               mask=alice_mask, 
               stopwords=stopwords)

wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## 2. 스타워즈

```python
text = open('datas/06_a_new_hope.txt').read()

text = text.replace("HAN", "Han")
text = text.replace("LUKE'S", "Luke")

mask = np.array(Image.open('datas/06_stormtrooper_mask.png'))
```

```python
print(text)
```

```python
stopwords.add('int')
stopwords.add('ext')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(max_words=1000,  # 표현할 최대 단어 갯수
               mask=mask,
               margin=10,
               random_state=1,
               stopwords=stopwords).generate(text)

default_colors = wc.to_array()
```

```python
import random

def grey_color_func(
        word, font_size, position, orientation, random_state=None, **kwargs) :
    return "hsl(0, 0%%, %d%%)" % random.randint(0, 100)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3))
plt.axis('off')
plt.show()
```
